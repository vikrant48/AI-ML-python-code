```python
stream = client.chat.completions.create(
    model='llama-3.1-8b-instant',
    temperature=0.3,
    stream=True,
    messages=[
        {'role' : 'user', 'content': 'Explain Stream in LLM'}
    ]
)
```

```python
print(stream)
```

```python
for chunk in stream:
    delta = chunk.choices[0].delta.content or ''
    print(delta, end='', flush=True)   # Print each token as it arrives
```

```python
import fitz  # PyMuPDF
import json

def extract_text_from_pdf(file_path):
    text = ""

    with fitz.open(file_path) as doc:
        for page in doc:
            text += page.get_text()

    return text
```

```python
resume_text = extract_text_from_pdf("resume.pdf")
jd_text = extract_text_from_pdf("JD.pdf")
print(resume_text)
print('-------------------------------------------------------------------------------------')
print(jd_text)
```

```python
SYSTEM = '''
You are an expert technical recruiter with 10 years of experience.
TASK: Evaluate resume–job description fit.
OUTPUT FORMAT (strict JSON):
{
  "match_score": 0-100,
  "match_reasons": ["reason1", "reason2"],
  "missing_skills": ["skill1", "skill2"],
  "recommendation": "Strong Match | Moderate Match | Weak Match"
}
RULES:- Be objective, cite specific skills- Score based on must-have vs nice-to-have- Return ONLY valid JSON, no extra text
'''
```

```python
def score_resume(resume_text, jd_text):
    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        temperature=0.3,
        messages=[
            {"role": "system", "content": SYSTEM},
            {
                "role": "user",
                "content": f"RESUME:\n{resume_text}\n\nJOB DESC:\n{jd_text}"
            }
        ]
    )

    return json.loads(response.choices[0].message.content)
```

```python
result = score_resume(resume_text, jd_text)
print(result)
print(json.dumps(result, indent=4))
```
